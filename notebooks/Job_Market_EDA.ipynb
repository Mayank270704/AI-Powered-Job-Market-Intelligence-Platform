```python
top_jobs = jobs['title'].value_counts().head(20)

print(top_jobs)
```

```python
import matplotlib.pyplot as plt

top_jobs = jobs['title'].value_counts().head(20)

plt.figure(figsize=(12,6))
top_jobs.plot(kind='bar')
plt.title("Top 20 Most Common Job Titles")
plt.xlabel("Job Title")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt

top_roles = jobs['title'].value_counts().head(15)

plt.figure(figsize=(12,6))
top_roles.plot(kind='bar')

plt.title('Top 15 Most In-Demand Job Roles')
plt.xlabel('Job Title')
plt.ylabel('Number of Openings')

plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

```python
jobs['formatted_work_type'].value_counts().plot(
    kind='pie',
    autopct='%1.1f%%',
    figsize=(8,8)
)

plt.title('Work Type Distribution')
plt.ylabel('')
plt.show()
```

```python
keywords = [
    'Data Analyst',
    'Data Scientist',
    'Machine Learning',
    'ML Engineer',
    'AI',
    'Business Analyst'
]

for keyword in keywords:
    count = jobs['title'].str.contains(
        keyword,
        case=False,
        na=False
    ).sum()

    print(f"{keyword}: {count}")
```

```python
analytics_jobs = jobs[
    jobs['title'].str.contains(
        'Data|Analyst|Scientist|Machine Learning|AI|Engineer',
        case=False,
        na=False
    )
]

analytics_jobs['title'].value_counts().head(15).plot(
    kind='bar',
    figsize=(12,6)
)

plt.title("Top Analytics & AI Roles")
plt.xlabel("Role")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
companies = pd.read_csv("../data/raw/companies.csv")

companies.head()
```

```python
print("Total Companies:", companies.shape[0])
```

```python
companies[['company_id','name']].head()
```

```python
merged = jobs.merge(
    companies[['company_id','name']],
    on='company_id',
    how='left'
)

merged[['title','name']].head()
```

```python
top_companies = merged['name'].value_counts().head(15)

print(top_companies)
```

```python
import matplotlib.pyplot as plt

top_companies.plot(
    kind='bar',
    figsize=(12,6)
)

plt.title("Top Hiring Companies")
plt.xlabel("Company")
plt.ylabel("Number of Job Posts")

plt.xticks(rotation=60)

plt.tight_layout()
plt.show()
```
